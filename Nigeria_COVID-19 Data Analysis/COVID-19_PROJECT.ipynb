```python
fig, axes=plt.subplots(1,2)
sns.regplot(x=np.log(ncdc_ext["No. of Cases (Lab Confirmed)"]), y="Population Density", data=ncdc_ext, ax=axes[0])
sns.regplot(x=np.log(ncdc_ext["No. of Deaths"]), y="Population Density", data=ncdc_ext, ax=axes[1])
```

An increase in the population density influences an increase in the number of confirmed cases and deaths.

```python
fig, axes=plt.subplots(1,2)
sns.regplot(x=np.log(ncdc_ext["No. of Cases (Lab Confirmed)"]), y="Socio-Economic", data=ncdc_ext, ax=axes[0])
sns.regplot(x=np.log(ncdc_ext["No. of Deaths"]), y="Socio-Economic", data=ncdc_ext, ax=axes[1])
```

A decrease in the Socio-Economic index influences an increase in the number of confirmed cases and deaths.

```python
fig, axes=plt.subplots(1,2)
sns.regplot(x=np.log(ncdc_ext["No. of Cases (Lab Confirmed)"]), y="Epidemiological", data=ncdc_ext, ax=axes[0])
sns.regplot(x=np.log(ncdc_ext["No. of Deaths"]), y="Epidemiological", data=ncdc_ext, ax=axes[1])
```

The higher the Epidemiological index, the higher the number of confirmed cases and deaths.

```python
fig, axes=plt.subplots(1,2)
sns.regplot(x=np.log(ncdc_ext["No. of Cases (Lab Confirmed)"]), y=" Transport Availability", data=ncdc_ext, ax=axes[0])
sns.regplot(x=np.log(ncdc_ext["No. of Deaths"]), y=" Transport Availability", data=ncdc_ext, ax=axes[1])
```

A decrease in transport availability influences an increase in the number of confirmed cases and deaths.

```python
gdp_melt=pd.melt(gdp, id_vars=['Year'], value_vars=['Q1', 'Q2', 'Q3', 'Q4'])
```

```python
gdp_plot=sns.barplot(x='Year', y='value', hue='variable', data=gdp_melt)
plt.legend(loc='lower left')
gdp_plot.axhline(y=15890000, color='r')
gdp_plot.set_title('GDP')
```

The GDP increaases towards the fourth quarter for each year
